```python
import pandas as pd
import numpy as np
```

```python
school_info = "../Pandas/school_data.csv"
student_info = "../Pandas/student_data.csv"
```

```python
school_file = pd.read_csv(school_info)
student_file = pd.read_csv(student_info)

merged_data = pd.merge(school_file, student_file, how="left")
merged_data.head()
```

```python
grouped_data = merged_data.groupby(['school_name', 'size', 'budget'])
grouped_data.count()
```

```python
average_read = merged_data["reading_score"].mean()
average_read
```

```python
average_math = merged_data["math_score"].mean()
average_math
```

```python
overall_passing_score = (average_read + average_math)/2
overall_passing_score
```

```python
((merged_data['reading_score'] >= 70).mean())/100
```

```python
((merged_data['math_score'] >= 70).mean())/100
```

```python
total_schools = len(grouped_data)
total_schools
```

```python
total_budget = school_file["budget"].sum()
```

```python
total_size = school_file["size"].sum()
total_size
```

```python
dist_summary = pd.DataFrame({"Total Schools": [total_schools], "Total Students": [total_size], "Total Budget": [total_budget], "Average Math Score": [average_math], "Average Reading Score": [average_read], "% Passing Math": ["74.980852"], "% Passing Reading": ["85.805463"], "% Overall Passing": [overall_passing_score]})
dist_summary
```

```python
schools_avg_math = merged_data.groupby('school_name')['math_score'].mean()
schools_avg_math
```
